# Exploratary Data Analysis

### Gender

```python
ax = sns.countplot(x = 'Gender',data = df , hue = "Gender" )

# we initialise this loop to get the value for each
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#checking sales on the basis of Gender
sales_gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y = 'Amount' , data = sales_gender, hue = "Gender" )
```

From above we can conclude that amount was spend by Females

### Age

```python
ax = sns.countplot(data = df, x = "Age Group" , hue = "Gender")
```

We can derive that there are more number of females in each age group

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, hue = "Age Group")
```

The most amount spend was in the age group of 26-35

### State

*Order Wise*

```python
# taking the top 10 performing states
orders_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
# setting this is important as the names should not overlap one another
sns.set(rc = {'figure.figsize':(16,5)})
sns.barplot(x = "State", y = 'Orders' , data = orders_state, hue = "State")
```

```python
orders_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
# setting this is important as the names should not overlap one another
sns.set(rc = {'figure.figsize':(16,5)})
sns.barplot(x = "State", y = 'Orders' , data = orders_state, hue = "State")
```

*Amount Wise*

```python
orders_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
# setting this is important as the names should not overlap one another
sns.set(rc = {'figure.figsize':(16,5)})
sns.barplot(x = "State", y = 'Amount' , data = orders_state, hue = "State")
```

*We can see from the following data that Most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively and it is the same for amount value of the orders*

### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status',hue = "Marital_Status")

sns.set(rc = {'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# checking on the basis of amount
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(6,5)})

sns.barplot(data = sales_marital, x = "Marital_Status" , y = "Amount", hue = "Gender")
```

*From above we can see that most of the buyers are women both married and unmarried*

### Occupation

```python
ax = sns.countplot(data = df, x = "Occupation", hue = "Occupation")
sns.set(rc = {'figure.figsize':(20,6)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(data = sales_occupation, x = "Occupation", y = "Amount", hue = "Occupation")
```

*From above we can see most of the buyers are from the IT Sector, Healthcare and Aviation Industry*

### Product Category


```python
ax = sns.countplot(x = "Product_Category", data = df, hue = "Product_Category")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_category = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.barplot(data = sales_category , x = "Product_Category" , y = "Amount", hue = "Product_Category")
```

*From the above 2 plots we can see that the most sold items belong to the Clothing Category but according to the sales we can clearly see that the most sales is clocked by Food Category*

*Top Selling Products*

```python
sale_product = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.barplot(x = 'Product_ID', y = 'Orders', data = sale_product, hue = 'Product_ID')
```

```python
# Alternate Ways
df.groupby("Product_ID")['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = "bar")
```

### Conclusion

*We can say that after all the analysis Married Women between the age group of 26-35 belonging to Uttar Pradesh, Maharashtra and Karnataka who are working in IT, Healthcare and Avaiation and more likely to buy products from the Clothing, Food and Electronics Category*
